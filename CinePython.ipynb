```python
class Product:
    def __init__(self, product_name, product_price):
        self.product_name = product_name
        self.product_price = product_price
    
    def __str__(self) -> str:
        return f'{self.product_name} - {self.product_price}'
```

```python
class Food_order:
    products=[]
    total_price_order = 0
    def __init__(self, restaurant_name, products ):
        self.restaurant_name = restaurant_name
        self.products = products
    def add_product(self, product):
        self.products.append(product)
    def delete_product(self, pos):
        self.products.pop(pos)
    def compute_total(self):
        for product in self.products:
            self.total_price_order += product.product_price
    def cancel_order(self):
        self.products.clear()
    def __str__(self) -> str:
        products_str = "Su pedido es:\n"
        for product in self.products:
            products_str += product.__str__() + "\n"
        return products_str + "el total del pedido es: " + str(self.total_price_order)
class Bill:
    price_per_hour = 50000
    total_price = 0
    food_orders = []
    def __init__(self, customer_name, customer_phone, rented_room_time, film_selected) :
        self.customer_name = customer_name
        self.customer_phone = customer_phone
        self.rented_room_time = rented_room_time
        self.film_selected = film_selected
        
    def add_order(self, food_order):
        self.food_orders.append(food_order)
    
    def compute_total(self):
        tot_for_orders = 0
        for order in self.food_orders:
            tot_for_orders += order.total_price_order
        self.total_price = tot_for_orders + (self.rented_room_time * self.price_per_hour)
    
    def __str__(self) -> str:
        res_part1 = """---Factura cine Python---\n
La reserva se encuentra a nombre de: {0}
El numero de contacto es: {1}
Tiempo de alquiler de la sala: {2}
Pelicula o serie a ver: {3}\n""".format(self.customer_name, self.customer_phone, self.rented_room_time,
                               self.film_selected)
        res_part2 = ""
        for food_order in self.food_orders:
            res_part2 += food_order.__str__() + "\n"
        return res_part1 + res_part2 + "el valor de la factura es: " + str(self.total_price) + "\n" +"--------------"
```
